### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación










```python
# Calcular ingresos por tienda
ingreso_tienda = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

# Mostrar resultados
print(f"Ingreso total Tienda 1: {ingreso_tienda:,.0f} COP")
print(f"Ingreso total Tienda 2: {ingreso_tienda2:,.0f} COP")
print(f"Ingreso total Tienda 3: {ingreso_tienda3:,.0f} COP")
print(f"Ingreso total Tienda 4: {ingreso_tienda4:,.0f} COP")
```

```python
import matplotlib.pyplot as plt

# Calcular ingresos
ingreso_tienda = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

# Datos para graficar
nombres = ["Tienda ", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [ingreso_tienda, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

# Crear gráfico
plt.figure(figsize=(8, 5))
plt.bar(nombres, ingresos, color='blue')
plt.title("Ingresos por Tienda", fontsize=14)
plt.ylabel("Ingresos (COP)", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

# 2. Ventas por categoría

```python
# Tienda
categorias_tienda1 = tienda["Categoría del Producto"].value_counts()
print("📍 Tienda 1 - Productos vendidos por categoría:")
print(categorias_tienda1, "\n")

# Tienda 2
categorias_tienda2 = tienda2["Categoría del Producto"].value_counts()
print("📍 Tienda 2 - Productos vendidos por categoría:")
print(categorias_tienda2, "\n")

# Tienda 3
categorias_tienda3 = tienda3["Categoría del Producto"].value_counts()
print("📍 Tienda 3 - Productos vendidos por categoría:")
print(categorias_tienda3, "\n")

# Tienda 4
categorias_tienda4 = tienda4["Categoría del Producto"].value_counts()
print("📍 Tienda 4 - Productos vendidos por categoría:")
print(categorias_tienda4)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Estilo visual
sns.set(style="whitegrid")

# Obtener conteo de categorías por tienda
categorias_tienda1 = tienda["Categoría del Producto"].value_counts()
categorias_tienda2 = tienda2["Categoría del Producto"].value_counts()
categorias_tienda3 = tienda3["Categoría del Producto"].value_counts()
categorias_tienda4 = tienda4["Categoría del Producto"].value_counts()

# Unificar categorías
todas_categorias = set(categorias_tienda1.index) | set(categorias_tienda2.index) | set(categorias_tienda3.index) | set(categorias_tienda4.index)

# Crear DataFrame
df = pd.DataFrame({
    'Categoría': list(todas_categorias),
    'Tienda 1': [categorias_tienda1.get(cat, 0) for cat in todas_categorias],
    'Tienda 2': [categorias_tienda2.get(cat, 0) for cat in todas_categorias],
    'Tienda 3': [categorias_tienda3.get(cat, 0) for cat in todas_categorias],
    'Tienda 4': [categorias_tienda4.get(cat, 0) for cat in todas_categorias]
})

# Transformar para graficar
df_melted = df.melt(id_vars='Categoría', var_name='Tienda', value_name='Cantidad')

# Crear gráfico
plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, x='Categoría', y='Cantidad', hue='Tienda', palette='Set2')
plt.title('📊 Productos Vendidos por Categoría en Cada Tienda', fontsize=14)
plt.xlabel('Categoría de Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45)
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio por tienda
calificacion_tienda = tienda["Calificación"].mean()
calificacion_tienda2 = tienda2["Calificación"].mean()
calificacion_tienda3 = tienda3["Calificación"].mean()
calificacion_tienda4 = tienda4["Calificación"].mean()

# Mostrar resultados
print(f"⭐ Tienda  - Calificación promedio: {calificacion_tienda1:.2f}")
print(f"⭐ Tienda 2 - Calificación promedio: {calificacion_tienda2:.2f}")
print(f"⭐ Tienda 3 - Calificación promedio: {calificacion_tienda3:.2f}")
print(f"⭐ Tienda 4 - Calificación promedio: {calificacion_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Productos más y menos vendidos por tienda

# Tienda 1
productos_tienda = tienda["Producto"].value_counts()
print("📍 Tienda 1")
print("Más vendido:", productos_tienda.idxmax(), "-", productos_tienda.max(), "ventas")
print("Menos vendido:", productos_tienda.idxmin(), "-", productos_tienda.min(), "venta(s)\n")

# Tienda 2
productos_tienda2 = tienda2["Producto"].value_counts()
print("📍 Tienda 2")
print("Más vendido:", productos_tienda2.idxmax(), "-", productos_tienda2.max(), "ventas")
print("Menos vendido:", productos_tienda2.idxmin(), "-", productos_tienda2.min(), "venta(s)\n")

# Tienda 3
productos_tienda3 = tienda3["Producto"].value_counts()
print("📍 Tienda 3")
print("Más vendido:", productos_tienda3.idxmax(), "-", productos_tienda3.max(), "ventas")
print("Menos vendido:", productos_tienda3.idxmin(), "-", productos_tienda3.min(), "venta(s)\n")

# Tienda 4
productos_tienda4 = tienda4["Producto"].value_counts()
print("📍 Tienda 4")
print("Más vendido:", productos_tienda4.idxmax(), "-", productos_tienda4.max(), "ventas")
print("Menos vendido:", productos_tienda4.idxmin(), "-", productos_tienda4.min(), "venta(s)")
```

```python
import seaborn as sns

# Conteo total por producto en cada tienda
productos1 = tienda["Producto"].value_counts()
productos2 = tienda2["Producto"].value_counts()
productos3 = tienda3["Producto"].value_counts()
productos4 = tienda4["Producto"].value_counts()

# Unificar productos
todos_productos = set(productos1.index) | set(productos2.index) | set(productos3.index) | set(productos4.index)

# Crear DataFrame
df_heat = pd.DataFrame({
    'Tienda 1': [productos1.get(p, 0) for p in todos_productos],
    'Tienda 2': [productos2.get(p, 0) for p in todos_productos],
    'Tienda 3': [productos3.get(p, 0) for p in todos_productos],
    'Tienda 4': [productos4.get(p, 0) for p in todos_productos]
}, index=list(todos_productos))

# Gráfico heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_heat, annot=True, fmt="d", cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap de Ventas por Producto y Tienda")
plt.xlabel("Tienda")
plt.ylabel("Producto")
plt.tight_layout()
plt.savefig("heatmap_ventas_productos_tiendas.png")
plt.show()
```

# 5. Envío promedio por tienda

```python
# Costo de envío promedio por tienda
costo_envio_tienda = tienda["Costo de envío"].mean()
costo_envio_tienda2 = tienda2["Costo de envío"].mean()
costo_envio_tienda3 = tienda3["Costo de envío"].mean()
costo_envio_tienda4 = tienda4["Costo de envío"].mean()

# Mostrar resultados
print(f"📦 Tienda  - Costo de envío promedio: R$ {costo_envio_tienda:.2f}")
print(f"📦 Tienda 2 - Costo de envío promedio: R$ {costo_envio_tienda2:.2f}")
print(f"📦 Tienda 3 - Costo de envío promedio: R$ {costo_envio_tienda3:.2f}")
print(f"📦 Tienda 4 - Costo de envío promedio: R$ {costo_envio_tienda4:.2f}")
```

```python
import seaborn as sns
import pandas as pd

# Crear DataFrame con todos los costos
df_envios = pd.DataFrame({
    "Tienda": ["Tienda 1"] * len(tienda) +
              ["Tienda 2"] * len(tienda2) +
              ["Tienda 3"] * len(tienda3) +
              ["Tienda 4"] * len(tienda4),
    "Costo de envío": pd.concat([
        tienda["Costo de envío"],
        tienda2["Costo de envío"],
        tienda3["Costo de envío"],
        tienda4["Costo de envío"]
    ], ignore_index=True)
})

# Gráfico boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_envios, x="Tienda", y="Costo de envío", palette="Set3")
plt.title("Distribución del Costo de Envío por Tienda")
plt.ylabel("Costo de Envío (R$)")
plt.tight_layout()
plt.savefig("grafico_caja.png")
plt.show()
```
